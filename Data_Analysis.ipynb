**_Exporatory Data Analysis_**

*Gender*

```python
sns.countplot(x = 'Gender', data= df)
```

```python
a=sns.countplot(x = 'Gender', data= df)

for bars in a.containers:
    a.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gender=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x ='Gender', y = 'Amount', data = sales_gender)
```

_Based on the above graph, It is evident that female customers dominate the sales volume and exhibit greater purchasing power than males. This implies that women not only constitute a larger share of the customer base but also spend more on average. This trend suggests that businesses should consider focusing their sales and marketing strategies more toward female shoppers to capitalize on this significant segment of the market._

**Age**

```python
a=sns.countplot(x = 'Age Group', data= df, hue='Gender')  

#'hue'= parameter is used to add a third dimension by grouping data points based on a specific categorical variable. 

for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Total Amount vs Age Group

sns.barplot(x ='Age Group', y = 'Amount', data = sales_age)
```

_From the above graph, it is evident that most of the buyers are females aged between 26-35 years old. This age group constitutes the largest portion of the customer base, reflecting their significant influence on purchasing decisions._

*State*

```python
# Total Order from top 15 States.

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)


sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(x ='State', y = 'Orders', data = sales_state)
```

```python
# Total Amount of 15 States

sales_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)


sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(x ='State', y = 'Amount', data = sales_amount)
```

_From the above graph, we can observe that most of the orders and total sales come from Uttar Pradesh, Maharashtra, and Karnataka, respectively. These states represent the highest contributions to sales, highlighting their importance as key markets for the business._

*Marital Status*

```python
a= sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(10,5)})
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x ='Marital_Status', y = 'Amount', data = sales_state, hue='Gender')
```

_From the above graph, it is clear that most of the buyers are married females, who also demonstrate higher purchasing power. This indicates that married women are significant consumers, reflecting their role in making purchasing decisions and suggesting targeted marketing strategies to cater to this demographic._

*Occupation*

```python
sns.set(rc={'figure.figsize':(20,5)})
a=sns.countplot(data = df,x = 'Occupation')
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(x ='Occupation', y = 'Amount', data = sales_state)
```

_From the above graph, it is evident that most of the buyers are employed in sectors such as IT, Healthcare, and Aviation. This correlation suggests that marketing strategies could be effectively tailored to target these specific professional groups, enhancing engagement and sales opportunities within these industries._

*Product Category*

```python
sns.set(rc={'figure.figsize':(40,35)})
a=sns.countplot(data = df,x = 'Product_Category')
for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)


sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(x ='Product_Category', y = 'Amount', data = sales_state)
```

_From the above graph, it is clear that the top five sold products are Food, Clothing & Apparel, Electronics & Gadgets, and Footwear & Shoes. This suggests significant consumer interest in these categories, indicating potential opportunities for targeted marketing and inventory management within these product lines._

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)


sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(x ='Product_ID', y = 'Orders', data = sales_state)
```
